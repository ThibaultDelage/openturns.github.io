# Create a white noise process

In this basic example we are going to build a white noise process by its distribution.

```python
from __future__ import print_function
import openturns as ot
import math as m
```

```python
# Define the distribution
myDist = ot.Normal()
```

```python
# Define the mesh
myMesh = ot.IntervalMesher([10]).build(ot.Interval([0.0], [1.0]))
```

```python
# Create the process:
process = ot.WhiteNoise(myDist, myMesh)
print(process)
```

```python
# Draw a sample
sample = process.getSample(6)
sample.drawMarginal(0)
```
