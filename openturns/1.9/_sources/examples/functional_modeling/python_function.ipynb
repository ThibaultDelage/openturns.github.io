# Create a Python function

In this basic example we are going to create a function from a Python function.

$$ f(x) = [x_0+x_1+x_2, (x_1-1)*e^{x_0} * x_2]$$

```python
from __future__ import print_function
import openturns as ot
import math as m

# define a function from R^3 to R^2
def regularFunc(x):
    x0, x1, x2 = x
    y0 = x0 + x1 + x2
    y1 = (x1 - 1.0) * m.exp(x0) * x2
    return [y0, y1]
```

```python
# create an OpenTURNS function from a regular Python function
function = ot.PythonFunction(3, 2, regularFunc)
x = [1.0, 2.0, 3.0]
print('x=', x, 'f(x)=', function(x))
```

```python
# draw y1 with x1=2.0, x2=1.0, x0 in [0, 2]
ot.ParametricFunction(function, [1, 2], [2.0, 1.0]).getMarginal(1).draw(0.0, 2.0)
```
