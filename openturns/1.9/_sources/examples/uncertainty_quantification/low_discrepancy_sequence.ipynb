# Create a low discrepancy sequence

In this basic example we are going to create a Sobol' sequence.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Create the sequence
dimension = 2
sequence = ot.SobolSequence(dimension)
```

```python
# Draw a sample
sample = sequence.generate(1000)
graph = ot.Graph('Sobol\' sequence', 'x1', 'x2', True, '')
cloud = ot.Cloud(sample)
graph.add(cloud)
graph
```
