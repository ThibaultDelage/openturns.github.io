# Visualize clouds

In this example we are going to draw clouds of points from a data sample.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Create data to visualize
dimension = 2
R = ot.CorrelationMatrix(dimension)
R[0, 1] = -0.8
size = 100
sample1 = ot.Normal([3.] * dimension, [2.] * dimension, R).getSample(size)
sample2 = ot.Normal([2.] * dimension, [3.] * dimension, R).getSample(size // 2)
```

```python
# Draw clouds
ot.VisualTest.DrawClouds(sample1, sample2)
```
