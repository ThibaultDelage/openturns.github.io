# Generate low discrepancy sequences

In this examples we are going to expose the available low discrepancy sequences that allow to approximate some integrals.
    

```python
from __future__ import print_function
import openturns as ot
import math as m
```

## 1. Sobol sequence

```python
sequence = ot.SobolSequence(2)
sample = sequence.generate(1024)
ot.Cloud(sample)
```

## 2. Halton sequence

```python
sequence = ot.HaltonSequence(2)
sample = sequence.generate(1024)
ot.Cloud(sample)
```

## 2. Reverse Halton sequence

```python
sequence = ot.ReverseHaltonSequence(2)
sample = sequence.generate(1024)
print('discrepancy=', ot.LowDiscrepancySequenceImplementation.ComputeStarDiscrepancy(sample))
ot.Cloud(sample)
```

## 3. Compare with uniform random sequence

```python
distribution = ot.ComposedDistribution([ot.Uniform(0.0, 1.0)]*2)
sample = distribution.getSample(1024)
print('discrepancy=', ot.LowDiscrepancySequenceImplementation.ComputeStarDiscrepancy(sample))
ot.Cloud(sample)
```
