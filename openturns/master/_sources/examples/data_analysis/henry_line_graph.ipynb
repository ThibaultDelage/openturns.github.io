# Normal fitting test using the Henry line

In this example we are going to perform a visual goodness-of-fit test for an univariate normal distribution.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Create data
ot.RandomGenerator.SetSeed(0)
distribution = ot.Normal(2.0, 0.5)
sample = distribution.getSample(100)
```

```python
# Draw Henry line plot
ot.VisualTest_DrawHenryLine(sample)
```
