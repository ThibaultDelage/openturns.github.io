# Visualize cobweb

In this example we are going to display a CobWeb plot from samples of X/Y data.

```python
from __future__ import print_function
import openturns as ot
```

```python
# Create data to visualize
size = 100
inputDimension = 6
inputSample = ot.Normal(inputDimension).getSample(size)
inputVar = [None]*inputDimension
for i in range(inputDimension):
    inputVar[i] = 'X' + str(i)
inputSample.setDescription(inputVar)
formula = ot.Description(1)
expression = ''
for i in range(inputDimension):
    if i > 0:
        expression += '+'
    expression += 'cos(' + str(i + 1) + '*' + inputVar[i] + ')'
formula[0] = expression
model = ot.SymbolicFunction(inputVar, formula)
outputSample = model(inputSample)
```

```python
# Draw CobWeb
graphCobweb = ot.VisualTest.DrawCobWeb(inputSample, outputSample, 1.0, 2.0, 'red', False)
graphCobweb.setLegendPosition('topleft')
graphCobweb
```
